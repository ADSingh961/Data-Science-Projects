1. Find out the number of unique characters in the sample conversation?

```python
unique_characters = []

text_file = open('conv.txt')
complete_text = text_file.readlines()
text_file.close()

for line in complete_text :
    if line == '\n':
        pass
    else :
        character = line.split(": ")[0] #WILL
        if character in unique_characters : 
            pass
        else :
            unique_characters.append(character)
            
print("Number of unique characters  : ", len(unique_characters)) 
print("The Unique Names are : ")

for name in unique_characters : 
    print(name, end = ', ')
```
